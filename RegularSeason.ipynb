```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
def add_game(games, resume):
    regular_game = 1
    for game in games:
        total = {}
    
        try:
            game['id']
            data = game.findNext('td').findNext('td').find('a').contents[0]
            age = data.findNext('td').contents[0]
            team = age.findNext('td').find('a').contents[0]
            opp = team.findNext('td').findNext('td').find('a').contents[0]
            result = opp.findNext('td').contents[0]
            started = result.findNext('td').contents[0]
            minutes = started.findNext('td').contents[0]

            fg = minutes.findNext('td').contents [0]
            fga = fg.findNext('td').contents[0]
            fgp = fga.findNext('td').contents[0]

            fg3 = fgp.findNext('td').contents[0]
            fg3a = fg3.findNext('td').contents[0]
            fg3p = fg3a.findNext('td').getText()

            ft = fg3a.findNext('td').findNext('td').contents[0]
            fta = ft.findNext('td').contents[0]
            ftp = fta.findNext('td').getText()

            orb = fta.findNext('td').findNext('td').contents[0]
            drb = orb.findNext('td').contents[0]
            totrb = drb.findNext('td').contents[0]
            ast = totrb.findNext('td').contents[0]
            stl = ast.findNext('td').contents[0]
            blk =stl.findNext('td').contents[0]
            tov = blk.findNext('td').contents[0]
            fouls = tov.findNext('td').contents[0]
            points = fouls.findNext('td').contents[0]

            total['game'] = regular_game
            total['data'] = data 
            total['age'] = age 
            total['team'] = team 
            total['opp'] = opp  
            total['result'] = result 
            total['started'] = started 
            total['minutes'] = minutes 

            total['fg'] = fg
            total['fga'] = fga
            total['fgp'] = fgp 

            total['fg3'] = fg3 
            total['fg3a'] = fg3a 
            total['fg3p'] = fg3p

            total['ft'] = ft
            total['fta'] = fta
            total['ftp'] = ftp

            total['orb'] = orb
            total['drb'] = drb 
            total['totrb'] = totrb 
            total['ast'] = ast  
            total['stl'] = stl
            total['blk'] = blk
            total['tov'] = tov
            total['fouls'] = fouls
            total['points'] = points


            regular_game += 1
            resume.append(total)
        except:
            game = game
        
```

```python
urls_jordan = []
resume_jordan = []
url = "https://www.basketball-reference.com/players/j/jordami01/gamelog/"
for i in range(1985, 1999):
    if i == 1994:
        continue
    else:
        new_url = url + str(i)
        urls_jordan.append(new_url)
urls_jordan.append(url + str(2002))
urls_jordan.append(url + str(2003))
urls_jordan
```

```python
headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/94.0.4606.61 Safari/537.36"}
```

```python
for url in urls_jordan:
    response = requests.get(url, headers=headers)
    text = response.text
    soup = BeautifulSoup(text, 'html.parser')
    

    list_games = soup.find('table', id ='pgl_basic')
    list_games = list_games.find('tbody')



    games = list_games.findAll('tr')

    add_game(games, resume_jordan)
```

```python
len(resume_jordan)
```

```python
urls_lebron = []
resume_lebron = []
url = "https://www.basketball-reference.com/players/j/jamesle01/gamelog/"
for i in range(2004, 2024):
        new_url = url + str(i)
        urls_lebron.append(new_url)
        
urls_lebron
```

```python
for url in urls_lebron:
    response = requests.get(url, headers=headers)
    text = response.text
    soup = BeautifulSoup(text, 'html.parser')
    

    list_games = soup.find('table', id ='pgl_basic')
    list_games = list_games.find('tbody')



    games = list_games.findAll('tr')

    add_game(games, resume_lebron)
```

```python
len(resume_lebron)
```

```python
jordan_games = pd.DataFrame(resume_jordan)
lebron_games = pd.DataFrame(resume_lebron)
```
